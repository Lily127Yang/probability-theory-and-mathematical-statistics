```python
import scipy
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
alpha=0.05
sum=0
iterations=100
for i in range(100):
    n=10
    samples=stats.norm.rvs(10,4,size=n)
    average=np.average(samples)
    plt.scatter(i,average)
    std=4
    left=10-scipy.stats.t.ppf(1-alpha/2,n-1)*std/(n**0.5)
    right=10+scipy.stats.t.ppf(1-alpha/2,n-1)*std/(n**0.5)
    plt.axhline(left)
    plt.axhline(right)
    if (left<=average)and (average<=right):
        sum+=1
print(sum/iterations)
print(1-sum/iterations)
print('left=',left)
print('right=',right)
#此时left和right是已知的定值
```
